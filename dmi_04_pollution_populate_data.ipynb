```python
import os
import pandas as pd
import numpy as np
import time
import warnings
# import matplotlib.pyplot as plt
# import seaborn as sns

# from sklearn.cluster import KMeans
# from sklearn.preprocessing import StandardScaler

# import folium
# from folium.plugins import MarkerCluster

from geopy.geocoders import Nominatim

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
file_path = './data/global_air_pollution_2.csv'  # Local path
```

```python
df = pd.read_csv(file_path)
print("Dataset shape:", df.shape)
```

```python
df.head(33)
```

```python
geolocator = Nominatim(user_agent="air_pollution_clustering")

# Ensure 'Latitude' and 'Longitude' columns exist in the DataFrame
if 'Latitude' not in df.columns:
    df['Latitude'] = None
if 'Longitude' not in df.columns:
    df['Longitude'] = None

# Define the output CSV file name
filename = './data/updated_air_pollution_dataset.csv'

# If the file doesn't exist, write the header using an empty dataframe slice
if not os.path.exists(filename):
    df.head(0).to_csv(filename, index=False)
    

def get_lat_lon(city, country):
    try:
        location = geolocator.geocode(f"{city}, {country}", timeout=10)
        if location:
            return location.latitude, location.longitude
        else:
            return np.nan, np.nan
    except Exception as e:
        return np.nan, np.nan

# if 'Latitude' not in df.columns or 'Longitude' not in df.columns:
latitudes = []
longitudes = []
for idx, row in df.iterrows():
    lat, lon = get_lat_lon(row['City'], row['Country'])
    print("found lat=", lat, "lon=", lon, "for", row['City'], row['Country'])
    latitudes.append(lat)
    longitudes.append(lon)

    # Only proceed if both lat and lon are not None
    if lat is not None and lon is not None:
        # Update the row with the new values
        row['Latitude'] = lat
        row['Longitude'] = lon

    # Append the updated row to the CSV file
    row.to_frame().T.to_csv(filename, mode='a', index=False, header=False)

    time.sleep(1)
df['Latitude'] = latitudes
df['Longitude'] = longitudes
```

```python
df.to_csv('./data/updated_airpollution_dataset.csv')
```
